```python
# import required libraries

import h5py
import numpy as np
import sys

from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split



LIB_PATH ="../../"
sys.path.append(LIB_PATH)
from utils_lib import *
```

```python
HDF5_DATASET_PATH ="../../h5-dataset/vehicle-type-dataset-SIZE224-train-dev-test-v2.hdf5"
print(HDF5_DATASET_PATH)
```

```python
class_label_string_length="S30"
```

```python
hf = h5py.File(HDF5_DATASET_PATH,"r")
```

```python
trainX = np.array(hf["trainX"]).astype("f8")
ascii_train_labels = np.array(hf["trainLabels"]).astype(class_label_string_length)
trainY = np.array(hf["trainY"]).astype("int")

devX = np.array(hf["devX"]).astype("f8")
ascii_dev_labels = np.array(hf["devLabels"]).astype(class_label_string_length)
devY = np.array(hf["devY"]).astype("int")

testX = np.array(hf["testX"]).astype("f8")
ascii_test_labels = np.array(hf["testLabels"]).astype(class_label_string_length)
testY = np.array(hf["testY"]).astype("int")

trainLabels = [n.decode("unicode_escape") for n in ascii_train_labels]
devLabels = [n.decode("unicode_escape") for n in ascii_dev_labels]
testLabels = [n.decode("unicode_escape") for n in ascii_test_labels]
```

```python
print(f"trainX.shape: {trainX.shape}")
print(f"trainY.shape: {trainY.shape}")
print(f"testX.shape: {testX.shape}")
print(f"testY.shape: {testY.shape}")
print(f"devX.shape: {devX.shape}")
print(f"devY.shape: {devY.shape}")
print(f"testX.shape: {testX.shape}")
print(f"testY.shape: {testY.shape}")
```
